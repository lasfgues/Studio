### Função alvo

$$ f(x) = \sqrt{2}$$

## Método de Newton

Como o método utiliza uma função variável em $x$ e a sua derivada, precisamos manipular a função ($\sqrt{2}$, que é uma constante) para que ela fique em função de uma variável $x$, só aí podemos aplicar o método.

$$ x = \sqrt{2} $$
$$ x^2 = (\sqrt{2})^2 $$
$$ x^2 = 2 $$
$$ x^2 - 2 = 0 $$
Logo,
$$ f(x) = x^2 - 2 $$
Cuja derivada,
$$ f'(x) = 2x$$

### Visualizando $f(x)$ e $f'(x)$.

#### Definindo as funções em Python.

```python
def f(x):
  return (x**2 - 2)

def fl(x):
  return 2 * x
```

#### Plotando os gráficos.

```python
x = np.linspace(-10, 10, 1000)
```

```python
fig, ax = plt.subplots(figsize = (15, 7))
ax.plot(x, f(x))
ax.plot(x, fl(x))
ax.grid()
fig.show()
```

### Aplicando o Método de Newton

$$ x_{k+1}=x_{k}-\dfrac{f(x_{k})}{f'(x_{k})}$$

```python
def newton(xi, epsilon, iterations):
  x = xi

  for i in range(iterations):
    x_newton = x - f(x) / fl(x)

    table.loc[i] = [x, f(x), fl(x), x_newton]

    if abs(x_newton - x) < epsilon:
      return x_newton, i

    x = x_newton

  return x, i
```

```python
table = pd.DataFrame(columns = ['x', 'f(x)', 'fl(x)', 'x_newton'])

x, i = newton(2, 0.00001, 100)
print(f'Solução: {x} \nIterações: {i}')

table
```
